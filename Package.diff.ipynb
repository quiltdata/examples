# Package diff
Allows you to narrow a manifest down to specific keys of interest.
See also [`Package.diff()`](https://docs.quiltdata.com/api-reference/package#Package.diff).

Data from [allencell repository](https://open.quiltdata.com/b/allencell/packages/aics/pipeline_integrated_single_cell).

```python
import quilt3
```

```python
pnew = quilt3.Package.browse(
    "aics/pipeline_integrated_single_cell",
    registry="s3://allencell",
    top_hash="a07796a429f2584a54abe2efe0f245f5794f21f530cb107cd4fd85e84082ff27"
)

pold = quilt3.Package.browse(
    "aics/pipeline_integrated_single_cell",
    registry="s3://allencell",
    top_hash="7b3f5e0b150d9492d9615a17e266a847ac7c194fb894a6cd19cea6e49e79dfb2"
)
```

Diff returns three lists: added, modified, deleted.

```python
pnew.diff(pold)
```
